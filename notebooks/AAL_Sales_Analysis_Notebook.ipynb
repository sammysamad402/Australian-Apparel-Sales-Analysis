```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import MinMaxScaler

sns.set_style("whitegrid")
plt.rcParams['figure.figsize']=(10,5)
```

```
df = pd.read_csv('AusApparalSales4thQrt2020.csv')
df.head()
```

```
print(df.shape)
df.info()
```

## Data Wrangling

```
print("Missing Values")
print(df.isna().sum())

print("\nNon Missing Values")
print(df.notna().sum())
```

```
print("Duplicate Records:", df.duplicated().sum())
df = df.drop_duplicates()
```

```
df.describe(include='all')
```

## Normalization

```
scaler = MinMaxScaler()

num_cols = df.select_dtypes(include=np.number).columns

df[num_cols] = scaler.fit_transform(df[num_cols])

df.head()
```

## Descriptive Statistics

```
sales_col = 'Sales'
unit_col = 'Unit'

print("Sales Statistics")
print("Mean:",df[sales_col].mean())
print("Median:",df[sales_col].median())
print("Mode:",df[sales_col].mode()[0])
print("Std:",df[sales_col].std())
```

```
print("Unit Statistics")
print("Mean:",df[unit_col].mean())
print("Median:",df[unit_col].median())
print("Mode:",df[unit_col].mode()[0])
print("Std:",df[unit_col].std())
```

```
sns.boxplot(data=df[[sales_col,unit_col]])
plt.title("Boxplot of Sales and Units")
plt.show()
```

## Group Analysis

```
group_sales = df.groupby('Group')['Sales'].sum().sort_values(ascending=False)
print(group_sales)

print("Highest Sales Group:",group_sales.idxmax())
print("Lowest Sales Group:",group_sales.idxmin())
```

```
sns.barplot(x=group_sales.index,y=group_sales.values)
plt.title("Group Wise Sales")
plt.xticks(rotation=45)
plt.show()
```

## State Analysis

```
state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False)
print(state_sales)

print("Highest State:",state_sales.idxmax())
print("Lowest State:",state_sales.idxmin())
```

```
sns.barplot(x=state_sales.index,y=state_sales.values)
plt.title("State Wise Sales")
plt.xticks(rotation=45)
plt.show()
```

## State vs Group Dashboard

```
pivot = pd.pivot_table(
    df,
    values='Sales',
    index='State',
    columns='Group',
    aggfunc='sum'
)

pivot.plot(kind='bar')
plt.title("State Wise Sales by Group")
plt.show()
```

## Date Analysis

```
df['Date'] = pd.to_datetime(df['Date'])
```

```
weekly_report = df.groupby(pd.Grouper(key='Date',freq='W'))['Sales'].sum()
weekly_report
```

```
monthly_report = df.groupby(pd.Grouper(key='Date',freq='M'))['Sales'].sum()
monthly_report
```

```
quarterly_report = df.groupby(pd.Grouper(key='Date',freq='Q'))['Sales'].sum()
quarterly_report
```

```
weekly_report.plot(marker='o')
plt.title("Weekly Sales Trend")
plt.show()
```

```
monthly_report.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.show()
```

## Time of Day Analysis

```
if 'Time' in df.columns:
    time_sales = df.groupby('Time')['Sales'].sum()
    print(time_sales)

    sns.barplot(x=time_sales.index,y=time_sales.values)
    plt.title('Sales by Time of Day')
    plt.show()
else:
    print("Time column not found in dataset")
```

## Distribution Plot

```
sns.histplot(df['Sales'], kde=True)
plt.title("Sales Distribution")
plt.show()
```
